**Q1.Create a vehicle class with an init method having instance variables as name_of_vehicle, max_speed
and average_of_vehicle.**

Certainly, here's a Python class definition for a `Vehicle` class with an `__init__` method that initializes the instance variables `name_of_vehicle`, `max_speed`, and `average_of_vehicle`:

```python
class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle
```

You can create instances of the `Vehicle` class by passing values for these attributes when you instantiate the class. For example:

```python
car = Vehicle("Car", 200, 30)
bike = Vehicle("Bike", 150, 40)

print(car.name_of_vehicle)  # Output: "Car"
print(car.max_speed)        # Output: 200
print(car.average_of_vehicle)  # Output: 30

print(bike.name_of_vehicle)  # Output: "Bike"
print(bike.max_speed)        # Output: 150
print(bike.average_of_vehicle)  # Output: 40
```

This code defines a `Vehicle` class with the specified instance variables and allows you to create instances of different vehicles with their respective attributes.

**Q2. Create a child class car from the vehicle class created in Que 1, which will inherit the vehicle class.
Create a method named seating_capacity which takes capacity as an argument and returns the name of
the vehicle and its seating capacity.**

Certainly, here's a Python code snippet that creates a child class `Car` from the `Vehicle` class and adds a `seating_capacity` method:

```python
class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle

class Car(Vehicle):
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        super().__init__(name_of_vehicle, max_speed, average_of_vehicle)

    def seating_capacity(self, capacity):
        return f"{self.name_of_vehicle} has a seating capacity of {capacity} people."

# Create an instance of the Car class
car = Car("Maruti", 200, 30)

# Call the seating_capacity method
capacity = 5
result = car.seating_capacity(capacity)

print(result)  # Output: "Maruti has a seating capacity of 5 people."
```

In this code:

1. The `Car` class is defined as a child class of the `Vehicle` class using inheritance.
2. The `__init__` method of the `Car` class calls the parent class (`Vehicle`) constructor using `super()` to initialize the inherited attributes.
3. The `seating_capacity` method is defined in the `Car` class, which takes the seating capacity as an argument and returns a string that includes the name of the vehicle and its seating capacity.
4. An instance of the `Car` class is created, and the `seating_capacity` method is called to get the seating capacity of the car.

**Q3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance.**

Multiple inheritance is a feature in object-oriented programming languages like Python, where a class can inherit attributes and methods from more than one parent class. This means that a class can have multiple base classes, and it inherits properties and behaviors from all of them.

Here's an example in Python that demonstrates multiple inheritance:

```python
# Parent class 1
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        pass

# Parent class 2
class Bird:
    def fly(self):
        pass

# Child class inheriting from both Animal and Bird
class Parrot(Animal, Bird):
    def speak(self):
        return f"{self.name} says 'Hello!'"

# Create an instance of the Parrot class
parrot = Parrot("Polly")

# Call methods from both parent classes
print(parrot.speak())  # Output: "Polly says 'Hello!'"
parrot.fly()  # Since Parrot inherits from Bird, it can also call the fly() method
```

In this example:

1. We define two parent classes, `Animal` and `Bird`, each with its own methods.
2. Then, we create a child class called `Parrot` that inherits from both `Animal` and `Bird`. This demonstrates multiple inheritance.
3. The `Parrot` class overrides the `speak` method from the `Animal` class.
4. We create an instance of the `Parrot` class and demonstrate that it can access and use methods from both parent classes, `Animal` and `Bird`.

Multiple inheritance can be a powerful tool, but it can also lead to complex class hierarchies, so it should be used carefully to maintain code clarity and avoid potential conflicts.

**Q4. What are getter and setter in python? Create a class and create a getter and a setter method in this
class.**

In Python, getter and setter methods are used to control access to an object's attributes (variables) in a class. They allow you to get (retrieve) and set (modify) the values of these attributes while providing a level of encapsulation and control over the data. Getter methods are used to retrieve the values, and setter methods are used to modify the values of the attributes.

Here's an example of a class with a getter and a setter method:

```python
class Student:
    def __init__(self, name, age):
        self._name = name  # Note the use of an underscore to indicate it's a "protected" attribute
        self._age = age

    # Getter method for 'name' attribute
    def get_name(self):
        return self._name

    # Setter method for 'name' attribute
    def set_name(self, name):
        if isinstance(name, str):
            self._name = name
        else:
            print("Name should be a string.")

    # Getter method for 'age' attribute
    def get_age(self):
        return self._age

    # Setter method for 'age' attribute
    def set_age(self, age):
        if isinstance(age, int) and age >= 0:
            self._age = age
        else:
            print("Age should be a positive integer.")

# Create an instance of the Student class
student = Student("Alice", 20)

# Get the current name and age
print("Name:", student.get_name())  # Output: "Name: Alice"
print("Age:", student.get_age())    # Output: "Age: 20"

# Set a new name and age using the setter methods
student.set_name("Bob")
student.set_age(22)

# Get the updated name and age
print("Name:", student.get_name())  # Output: "Name: Bob"
print("Age:", student.get_age())    # Output: "Age: 22"

# Try setting an invalid name and age
student.set_name(123)  # Output: "Name should be a string."
student.set_age(-5)    # Output: "Age should be a positive integer."
```

In this example, the `Student` class has attributes `_name` and `_age`, which are considered "protected" attributes (indicated by the single underscore convention). Getter methods (`get_name` and `get_age`) are used to retrieve the values, and setter methods (`set_name` and `set_age`) are used to set new values for these attributes. The setter methods include validation checks to ensure that the provided values are of the correct type and within valid ranges.

**Q5.What is method overriding in python? Write a python code to demonstrate method overriding.**

Method overriding in Python is a feature of object-oriented programming that allows a subclass to provide a specific implementation of a method that is already defined in its superclass. When a method in a subclass has the same name, parameters, and return type as a method in its superclass, it is said to override the superclass's method. Method overriding is used to provide a more specialized or customized behavior for a method in a subclass.

Here's an example in Python that demonstrates method overriding:

```python
class Animal:
    def speak(self):
        return "Animal speaks."

class Dog(Animal):
    def speak(self):
        return "Dog barks."

class Cat(Animal):
    def speak(self):
        return "Cat meows."

# Create instances of the Dog and Cat classes
dog = Dog()
cat = Cat()

# Call the speak method on instances of both classes
print(dog.speak())  # Output: "Dog barks."
print(cat.speak())  # Output: "Cat meows."
```

In this example:

1. We have a base class `Animal` with a `speak` method that provides a generic implementation.
2. We then create two subclasses, `Dog` and `Cat`, which inherit from the `Animal` class.
3. Both `Dog` and `Cat` classes override the `speak` method with their own specific implementations.

When you call the `speak` method on instances of `Dog` and `Cat`, they return the specialized behavior defined in their respective subclasses. This is an example of method overriding in Python, where the subclass provides its own version of a method with the same name as the one in the superclass.
