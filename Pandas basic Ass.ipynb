**Q1. Create a Pandas Series that contains the following data: 4, 8, 15, 16, 23, and 42. Then, print the series.**

Python using Pandas to create a series with the given data and print it:

```python
import pandas as pd

# Create a Pandas Series
data = [4, 8, 15, 16, 23, 42]
my_series = pd.Series(data)

# Print the Series
print(my_series)
```

This will output:

```
0     4
1     8
2    15
3    16
4    23
5    42
dtype: int64
```

Each number is associated with an index in the series. The indices in this case are 0, 1, 2, 3, 4, and 5.

**Q3. Create a Pandas DataFrame that contains the following data:**

<p>Name             Age     Gender</p>
<p>Alice            25      Female</p>
<p>Bob              30      Male</p>
<p>Claire           27      Female</p>

Then, print the DataFrame.

Certainly! You can create a Pandas DataFrame using the provided data as follows:

```python
import pandas as pd

# Create a dictionary with the data
data = {
    'Name': ['Alice', 'Bob', 'Claire'],
    'Age': [25, 30, 27],
    'Gender': ['Female', 'Male', 'Female']
}

# Create a Pandas DataFrame from the dictionary
df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```

In this DataFrame, each column corresponds to a different attribute (Name, Age, Gender), and each row represents a different person. The indices (0, 1, 2) are automatically assigned by Pandas.

**Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.**

In Pandas, a `DataFrame` is a two-dimensional, tabular data structure that consists of columns and rows. It is similar to a spreadsheet or SQL table, where each column can have a different data type. A `DataFrame` can be thought of as a container for one or more Pandas Series. Essentially, it's a way to organize and manipulate data in a structured format.

On the other hand, a Pandas `Series` is a one-dimensional labeled array capable of holding any data type. It can be considered as a single column of a DataFrame. Each element in a Series is assigned a label (an index), and you can access elements using these labels.

Let's illustrate the difference between a DataFrame and a Series with an example:

```python
import pandas as pd

# Creating a Series
ages = pd.Series([25, 30, 27], name='Age')

# Creating a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Claire'],
    'Age': [25, 30, 27],
    'Gender': ['Female', 'Male', 'Female']
}
df = pd.DataFrame(data)

# Printing the Series
print("Series:")
print(ages)
print("\n")

# Printing the DataFrame
print("DataFrame:")
print(df)
```

Output:
```
Series:
0    25
1    30
2    27
Name: Age, dtype: int64

DataFrame:
     Name  Age  Gender
0   Alice   25  Female
1     Bob   30    Male
2  Claire   27  Female
```

In this example, the `ages` variable is a Pandas Series representing the 'Age' column of the DataFrame. The DataFrame, `df`, contains three columns ('Name', 'Age', 'Gender'), and each column is a Pandas Series. The primary difference is that a DataFrame can hold multiple Series, each representing a column, while a Series is a single column of data.

**Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can
you give an example of when you might use one of these functions?**

Pandas provides a wide range of functions for manipulating and analyzing data in a DataFrame. Here are some common functions:

1. **head() and tail():** These functions are used to view the first or last n rows of a DataFrame, respectively.
    ```python
    df.head()  # View the first 5 rows of the DataFrame
    ```

2. **describe():** Provides summary statistics of numerical columns in the DataFrame.
    ```python
    df.describe()
    ```

3. **info():** Gives a concise summary of a DataFrame, including the data types and non-null values.
    ```python
    df.info()
    ```

4. **loc[] and iloc[]:** Used for selection by label (loc) or by index (iloc).
    ```python
    df.loc[0]   # Selects the row with label 0
    df.iloc[0]  # Selects the first row by index
    ```

5. **value_counts():** Returns a Series containing counts of unique values in a column.
    ```python
    df['Gender'].value_counts()
    ```

6. **groupby():** Groups the DataFrame using a mapper or by a Series of columns, and then applies a function to each group.
    ```python
    df.groupby('Gender')['Age'].mean()
    ```

7. **sort_values():** Sorts the DataFrame by the values along either axis.
    ```python
    df.sort_values(by='Age', ascending=False)
    ```

8. **fillna():** Fills missing values with a specified value or method.
    ```python
    df.fillna(0)  # Replace NaN values with 0
    ```

9. **drop():** Removes specified rows or columns from the DataFrame.
    ```python
    df.drop(['Age'], axis=1)  # Removes the 'Age' column
    ```

10. **apply():** Applies a function along the axis of the DataFrame.
    ```python
    df['Age'].apply(lambda x: x + 1)  # Adds 1 to each element in the 'Age' column
    ```

These are just a few examples, and there are many more functions available in Pandas for various data manipulation tasks. The choice of function depends on the specific task you want to perform on your data.

**Q6. Which of the following is mutable in nature Series, DataFrame, Panel?**

In Pandas:

- **Series:** It is mutable. You can modify the values in a Series after it has been created.

- **DataFrame:** It is also mutable. You can modify the values in a DataFrame, add or remove columns, and perform various operations to change its structure.

- **Panel:** Panels have been deprecated in recent versions of Pandas, and their use is discouraged. Instead, multi-index DataFrames are recommended. If you are working with newer versions of Pandas, you'll typically be dealing with Series and DataFrames rather than Panels.

In summary, both Series and DataFrames are mutable, but Panels are deprecated, and you should use multi-index DataFrames instead.

**Q7. Create a DataFrame using multiple Series. Explain with an example.**

Certainly! You can create a DataFrame using multiple Series by combining them into a dictionary where the keys are the column names, and the values are the Series. Here's an example:

```python
import pandas as pd

# Create multiple Pandas Series
names = pd.Series(['Alice', 'Bob', 'Claire'], name='Name')
ages = pd.Series([25, 30, 27], name='Age')
genders = pd.Series(['Female', 'Male', 'Female'], name='Gender')

# Combine the Series into a DataFrame
data = {
    'Name': names,
    'Age': ages,
    'Gender': genders
}

df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```

Output:
```
     Name  Age  Gender
0   Alice   25  Female
1     Bob   30    Male
2  Claire   27  Female
```

In this example, we created three Pandas Series (`names`, `ages`, and `genders`), each representing a column of the DataFrame. Then, we combined these Series into a dictionary (`data`), where the keys are the column names, and the values are the corresponding Series. Finally, we used the `pd.DataFrame()` constructor to create the DataFrame (`df`) from the dictionary.

The resulting DataFrame has three columns: 'Name', 'Age', and 'Gender', with the data from the respective Series.
